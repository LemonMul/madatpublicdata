```python
# 필요한 패키지 임포트
import requests
import pandas as pd
```

# 데이터 로드

```python
# API 요청

# API 요청 # 17초 소요
start_point = 1
end_point = 1000 # 최대 1000개까지만 호출 할 수 있음
api_key = '57524f76506d656e3732636a52457a'


url = f'http://openAPI.seoul.go.kr:8088/{api_key}/json/SearchParkInfoService/{start_point}/{end_point}/'

data = requests.get(url).json()
data.keys() # ['SearchParkInfoService']

df = pd.DataFrame(data['SearchParkInfoService']['row'])

df.tail()
```

```python
# 컬럼명 한글로 바꾸기

column_mapping = {
    "P_IDX": "연번 (공원번호)",
    "P_PARK": "공원명",
    "P_LIST_CONTENT": "공원개요",
    "AREA": "면적",
    "OPEN_DT": "개원일",
    "MAIN_EQUIP": "주요시설",
    "MAIN_PLANTS": "주요식물",
    "GUIDANCE": "안내도",
    "VISIT_ROAD": "오시는길",
    "USE_REFER": "이용시참고사항",
    "P_IMG": "이미지",
    "P_ZONE": "지역",
    "P_ADDR": "공원주소",
    "P_NAME": "관리부서",
    "P_ADMINTEL": "전화번호",
    "G_LONGITUDE": "X좌표(GRS80TM)",
    "G_LATITUDE": "Y좌표(GRS80TM)",
    "LONGITUDE": "X좌표(WGS84)",
    "LATITUDE": "Y좌표(WGS84)",
    "TEMPLATE_URL": "바로가기"
}

# 컬럼명 변경
df.rename(columns=column_mapping, inplace=True)
```

```python
cleaned_df = df[['공원명','공원주소','전화번호']]
```
