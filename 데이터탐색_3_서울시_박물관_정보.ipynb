```python
import pandas as pd
```

# 데이터 로드

```python
# csv 파일 불러오기
df = pd.read_csv('서울특별시_박물관미술관정보_20230323.csv', encoding='cp949')
```

```python
df.head()
```

```python
df.columns
```

```python
# '주소' 컬럼 생성 # 도로명주소가 결측값이면 지번주소사용
import numpy as np
df['주소'] = np.where(df['소재지도로명주소'].isna() | (df['소재지도로명주소'] == '-') | (df['소재지도로명주소'].str.strip() == ''), df['소재지지번주소'], df['소재지도로명주소'])
```

```python
df.columns
```

```python
cleaned_df = df[['시설명','주소','운영기관전화번호']]
```

# csv 파일로 변환

```python
# tsv 파일로 변환 (콤마가 아닌 탭으로 구분)

df.to_csv('seoul_museums.tsv', sep='\t', index=False, encoding='cp949')
cleaned_df.to_csv('cleaned_seoul_museums.tsv', sep='\t', index=False, encoding='cp949')
```
