# 데이터 로드

```python
# 필요한 패키지 임포트
import requests
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# API 요청 # 17초 소요
start_point = 1
end_point = 1000
url1 = f'http://openAPI.seoul.go.kr:8088/57524f76506d656e3732636a52457a/json/ListNecessariesPricesService/{start_point}/{end_point}'
requests.get(url1)

# json 파일로 자료 요청
json1 = requests.get(url1).json()
```

```python
# json 파일 확인 # 우리가 가져올 것은 ['ListNecessariesPricesService']키의 밸류 'row'를 키로하는 밸류 값들 = 우리 dataset의 컬럼
# json1 # 14개의 컬럼을 2개의 열만 가져옴
json1.keys() # ['ListNecessariesPricesService']
json1['ListNecessariesPricesService'].keys() # ['row']
data = json1['ListNecessariesPricesService']['row'] # 우리가 원하는 dataset
data # 리스트 안에 딕셔너리가 있는 것을 확인
# # type(data) # list

df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
# 결측치 처리
# 결측치가 있는지 확인
# print("결측치가 있는 행 개수:", df.isnull().sum()) # 결측치 없음

# 'A_PRICE'의 자료형을 정수형으로 바꿈
df['A_PRICE'] = df['A_PRICE'].astype(int)

# 결측치는 없지만 0이 몇 개인지 확인
df.sort_values(by='A_PRICE', ascending=True).head() # 있음
# 'A_PRICE'의 값이 0 이면 행을 날림
df = df[df['A_PRICE'] != 0]

print("결측치가 제거된 후 행 개수:", len(df))
print(df)
```

## 가장 많이 나온 항목은 감자

```python
# A_NAME 컬럼에서 띄어쓰기를 기준으로 앞 값만 추출 (수량을 제외하고 제품 이름만 추출해서 새컬럼 만들기)
df['A_NAME_PLUS'] = df['A_NAME'].str.split().str[0]
df['A_NAME_PLUS']
df['A_NAME_PLUS'].value_counts() # 감자가 가장 많이 나온 항목 # 다들 비슷한 값이라 의미는 없음
```

## 감자가 제일 비쌌던 날, 제일 저렴했던 날

```python
potato_max = df[df['A_NAME_PLUS'] == "감자"].sort_values(by='A_PRICE', ascending=False).head(1)
potato_min = df[df['A_NAME_PLUS'] == "감자"].sort_values(by='A_PRICE', ascending=False).tail(1)
pd.concat([potato_max,potato_min]) # 하루 만에 4천원 차이
```

## 가장 비쌌던 항목은 쌀, 고춧가루(국산), 수박

```python
df.columns  # 역시 수량이 무시된 값이라 딱히 의미는 없음
exp_item = df.groupby('A_NAME_PLUS')['A_PRICE'].max().sort_values(ascending=False)
display(exp_item.head())
```

## 쌀이 제일 비쌌던 날, 제일 저렴했던 날

```python
rice_max = df[df['A_NAME_PLUS'] == "쌀"].sort_values(by='A_PRICE', ascending=False).head(1)
rice_min = df[df['A_NAME_PLUS'] == "쌀"].sort_values(by='A_PRICE', ascending=False).tail(1)
pd.concat([rice_max,rice_min]) # 하루 만에 48,200원 차이
```
