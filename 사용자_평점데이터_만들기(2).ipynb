```python
import json
import pandas as pd

# 파일 열기 및 JSON 데이터 로드
with open('reviews_par.json', 'r', encoding='utf-8') as file:
    parData = json.load(file)

# 모든 리뷰 데이터를 하나의 DataFrame으로 통합
rows = []
for place, reviews in parData.items():
    for review in reviews:
        row = review + [place]  # place_name을 리스트에 추가
        rows.append(row)

# DataFrame 생성
parDf = pd.DataFrame(rows, columns=['User Name', 'Rating', 'URL', 'Review Text', 'Place Name'])

display(parDf.tail(3))
```

유사도로 평점계산

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics.pairwise import cosine_similarity



# 데이터 분할
ratings = df[['User Id', 'Place Id', 'Rating']]
train_data, test_data = train_test_split(ratings, test_size=0.2, random_state=42)

# 훈련 세트와 테스트 세트로 평점 행렬 생성
train_matrix = train_data.pivot_table(index='User Id', columns='Place Id', values='Rating').fillna(0)
test_matrix = test_data.pivot_table(index='User Id', columns='Place Id', values='Rating').fillna(0)

# 사용자 유사도 계산
user_similarity_train = cosine_similarity(train_matrix)
user_similarity_train_df = pd.DataFrame(user_similarity_train, index=train_matrix.index, columns=train_matrix.index)

# 예측 함수
# 테스트 데이터의 실제 사용자-장소 행렬을 생성할 때, 인덱싱 오류를 방지하기 위한 처리
def filter_predictions(predicted_ratings, test_user_ids, test_place_ids):
    # 훈련 세트에 있는 사용자 ID와 장소 ID에 대해서만 예측값을 추출
    test_user_mask = np.isin(test_user_ids, predicted_ratings.index)
    test_place_mask = np.isin(test_place_ids, predicted_ratings.columns)

    filtered_user_ids = test_user_ids[test_user_mask]
    filtered_place_ids = test_place_ids[test_place_mask]

    # 예측 행렬에서 필터링된 사용자 ID와 장소 ID를 사용하여 예측값을 추출
    predicted_test_ratings = predicted_ratings.loc[filtered_user_ids, filtered_place_ids]

    return predicted_test_ratings, filtered_user_ids, filtered_place_ids

# 예측 행렬을 DataFrame으로 변환
predicted_ratings_df = pd.DataFrame(predicted_ratings, index=train_matrix.index, columns=train_matrix.columns)

# 필터링된 예측 평점과 사용자 ID, 장소 ID 추출
predicted_test_ratings, filtered_user_ids, filtered_place_ids = filter_predictions(predicted_ratings_df, test_user_ids, test_place_ids)

# 실제 평점 행렬 생성 및 인덱싱 오류 방지
real_test_ratings = test_matrix.loc[filtered_user_ids, filtered_place_ids]

# RMSE 계산
rmse = np.sqrt(mean_squared_error(real_test_ratings, predicted_test_ratings))
print(f"RMSE: {rmse}")
```

```python
# 사용자 11번에 대한 평점 예측
def predict_user_ratings(user_id, user_similarity_df, ratings_matrix):
    if user_id in user_similarity_df.index:
        # 해당 사용자와 다른 사용자들 간의 유사도
        user_similarity_scores = user_similarity_df.loc[user_id]
        # 사용자의 평점 행렬
        user_ratings = ratings_matrix.loc[user_id]

        # 사용자가 평가하지 않은 장소에 대한 예측 평점 계산
        weighted_sum = np.dot(user_similarity_scores, ratings_matrix)
        sim_sum = np.array([np.abs(user_similarity_scores).sum()])
        pred_ratings = np.divide(weighted_sum, sim_sum, where=sim_sum!=0)  # 0으로 나누는 것 방지

        # 스케일링 복원
        min_rating = ratings_matrix.min().min()
        max_rating = ratings_matrix.max().max()
        pred_ratings = (pred_ratings * (max_rating - min_rating)) + min_rating

        # 이미 평가한 장소는 제외하고 예측
        missing_ratings = ratings_matrix.columns.isin(user_ratings[user_ratings > 0].index)
        predictions = pd.Series(pred_ratings, index=ratings_matrix.columns)
        return predictions[~missing_ratings]
    else:
        return "User ID not found in the similarity matrix"

# 예측 평점 출력
user_11_predictions = predict_user_ratings(17, user_similarity_train_df, train_matrix)
print(user_11_predictions)


user_11_predictions.sort_values(ascending=False)
```
